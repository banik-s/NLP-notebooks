## Part 1: Sentence & Word Tokenization in spaCy
spaCy uses a **statistical model** (dependency parser) to detect sentence boundaries. It correctly treats `Dr.` as an abbreviation (not a sentence end) because the model has learned this from data.

Load the pre-trained `en_core_web_sm` model and pass the text through it to create a `Doc` object — spaCy's central data structure containing tokens, sentences, entities, and more.

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("Dr. Strange loves pav bhaji of mumbai. Hulk loves chat of delhi")
```

`doc.sents` is a generator over detected sentences. spaCy correctly splits into **2 sentences**, recognizing `Dr.` as an honorific rather than a sentence boundary — a key advantage over simple rule-based splitting.

```python
for sentence in doc.sents:
    print(sentence)
```

Iterating over sentences and then each token gives a two-level view: sentence → word. Notice punctuation (`.`) is split as its own token — this is deliberate; it preserves meaning and enables downstream POS/NER tasks.

```python
for sentence in doc.sents:
    for word in sentence:
        print(word)
```

## Part 2: Sentence & Word Tokenization in NLTK
NLTK's `sent_tokenize` uses **Punkt**, a rule-based sentence tokenizer. Download the `punkt_tab` resource (required once) before calling it.

`nltk.download('punkt_tab')` fetches the pre-trained Punkt model data needed for sentence tokenization. It is downloaded to `~/nltk_data` and cached for future sessions.

```python
from nltk.tokenize import sent_tokenize
import nltk
nltk.download('punkt_tab')
```

 **Key difference:** NLTK's Punkt tokenizer incorrectly treats `Dr.` as a sentence boundary, producing **3 segments** (`['Dr.', 'Strange loves...', 'Hulk loves...']`). spaCy correctly returns **2**.

```python
sent_tokenize("Dr. Strange loves pav bhaji of mumbai. Hulk loves chat of delhi")
```

`word_tokenize` splits text into individual tokens including punctuation. Internally it uses the TreebankWordTokenizer (Penn Treebank rules).

```python
from nltk.tokenize import word_tokenize
```

NLTK splits `Dr.` → `['Dr', '.']` (separates the period), while spaCy keeps `Dr.` as one token. Neither is wrong — the difference reflects the tokenization philosophy (linguistic vs. rule-based).

```python
word_tokenize("Dr. Strange loves pav bhaji of mumbai. Hulk loves chat of delhi")
```

## Summary: NLTK vs spaCy at a Glance

| Dimension | NLTK | spaCy |
|---|---|---|
| **Philosophy** | Research / teaching toolkit | Production-ready NLP |
| **Ease of use** | Manual pipeline assembly | Pre-built, opinionated pipeline |
| **Speed** | Slower (pure Python) | Faster (Cython-compiled) |
| **Pre-trained models** | No built-in DL models | Yes — NER, POS, parsing |
| **Tokenization** | Rule-based (Punkt/Treebank) | Statistical + rule-based |
| **NER** | Requires Stanford NER / manual | Built-in, state-of-the-art |
| **Best for** | Learning, experimentation | Real-world NLP apps |

> **Rule of thumb:** Use NLTK to understand NLP concepts; use spaCy to build NLP systems.
